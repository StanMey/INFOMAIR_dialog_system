Pre-process the data

```python
import csv
import pandas as pd

from google.colab import drive
```

```python
enable_gdrive = True   # Switch to indicate whether Google Drive should be mounted.

if enable_gdrive:
    drive.mount('/content/drive')
```

```python
# Load the dataset
with open("/content/drive/MyDrive/Methods in AI research/datasets/dialog_acts.dat") as f:
  # convert to lower and remove the \n
  data = list(map(lambda x: x.lower().replace("\n", ""), f.readlines()))

data[:10]
```

```python
# extract the dialog acts from the string
data = [(x.split(" ")[0], " ".join(x.split(" ")[1:])) for x in data]
data[:10]
```

```python
# split into training and test part
training_size = 0.85
data_split = round(training_size * len(data))
# print(len(data), data_split)

# make the split based on the percentage and size of the dataset
train_data = data[:data_split]
test_data = data[data_split:]
```

```python
# convert to pandas for easy storing later on
header = ["dialog_act", "utterance_content"]
train_path = "/content/drive/MyDrive/Methods in AI research/datasets/training_dialog"
test_path = "/content/drive/MyDrive/Methods in AI research/datasets/test_dialog"

train_df = pd.DataFrame(train_data, columns=header)
test_df = pd.DataFrame(test_data, columns=header)
```

Check for duplicates

```python
data_df = pd.DataFrame(data, columns=header)
print(f"size dataset: {len(data_df)}\nDuplicates in dataset: {data_df.duplicated().sum()}")
```

```python
print(f"size train_df: {len(train_df)}\nDuplicates in train df: {train_df.duplicated().sum()}\n\nsize test_df: {len(test_df)}\nDuplicates in test df: {test_df.duplicated().sum()}")
```

```python
# save to pickle
train_df.to_pickle(f"{train_path}.pkl")
test_df.to_pickle(f"{test_path}.pkl")

# save to csv
train_df.to_csv(f"{train_path}.csv", encoding='utf-8')
test_df.to_csv(f"{test_path}.csv", encoding='utf-8')
```
